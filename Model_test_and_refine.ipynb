```python
## starting model refinement and testing in this notebook 

# Import our dependencies
import pandas as pd



# Read in our ramen data
num_all_homes_df_transformed = pd.read_csv("trans_with_state1HOT_Zhvi_city.csv")
# Read in our ramen data
num_all_homes_df_transformed
# # Print out the Country value counts
# country_counts = ramen_df.Country.value_counts()
# country_counts
```

```python
## check that KNN imputer worked 

    
for column in num_all_homes_df_transformed.columns:
    print(f"column {column} has {num_all_homes_df_transformed[column].isnull().sum()} null values")
```

```python
num_all_homes_df_transformed['states_Types_Cat'].value_counts()
```

```python
# Define features set -- using 2020-03 as y for now 
X = num_all_homes_df_transformed.copy()
X = X.drop("2020-03-31", axis=1)
X.head()
```

```python
# Define target vector y 
y = num_all_homes_df_transformed["2020-03-31"].values.reshape(-1, 1)
y[:5]
```

```python
print(y.shape)
print(type(y))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# ## instead of train test will just compare x and y with cross-validation 

# Creating StandardScaler instance
scaler = StandardScaler()
# Fitting Standard Scaller
X_scaler = scaler.fit(X)
# Scaling data
X_scaled = X_scaler.transform(X)
```

```python
## fitting scaled x and y to linear regression model 

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_scaled,y)

```

```python
## create small test set

some_y = y[0:200]
some_x = X_scaled[0:200]
```

```python
## make predictions with trained model 

lin_reg.predict(some_x)
```

```python
print("actual :",list(some_y))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
import os

housing_predictions = lin_reg.predict(X_scaled)
lin_mse = mean_squared_error(y, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(y, housing_predictions)
lin_mae
```

```python
## mean absolute error large on lin reg 
```

```python
## trying decision tree regressor 

from sklearn.tree import DecisionTreeRegressor
## create instance 
tree_reg = DecisionTreeRegressor(random_state=42)

# fit to data 
tree_reg.fit(X_scaled, y)
```

```python
## over fitting initially as it has seen all th data 

housing_predictions = tree_reg.predict(X_scaled)
tree_mse = mean_squared_error(y, housing_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, X_scaled, y,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse_scores)
```

```python
### check cross validation scores for linear regressor 

lin_scores = cross_val_score(lin_reg, X_scaled, y,
                             scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=10, random_state=42, max_depth=3)
forest_reg.fit(X_scaled, y)
```

```python
lin_scores = cross_val_score(forest_reg, X_scaled, y,
                             scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
##at the moment the lin_regressor has better mean square error than the tree regressor or random forest regressor

## more tunning of hyperparameters could help 

## will create predicitions csv with lin_reg for second deliverable 
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python

param_distribs = {
        'n_estimators': randint(low=1, high=10),
        'max_features': randint(low=1, high=200),
    }

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(X_scaled, y)
```

```python
## results of random cv of random forest regressor 

cvres = rnd_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
## taking best model from CV 

final_model = rnd_search.best_estimator_
```

```python
## the hpyerparameters for the best estimator 

rnd_search.best_estimator_
```

```python
final_predictions = final_model.predict(X_scaled)

final_mse = mean_squared_error(y, final_predictions)
final_rmse = np.sqrt(final_mse)
```

```python
## real mean square error less than 10 thousand , not bad compared to others, 

final_rmse
```
